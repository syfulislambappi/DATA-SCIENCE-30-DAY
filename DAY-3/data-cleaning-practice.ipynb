```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("loan.csv")
```

```python
# See the overview of dataset
df.head()
```

```python
# Get the number of rows and columns of data
df.shape
```

```python
# Get the number of missing values in terms of columns
df.isnull().sum()
```

```python
# Get the percentaqge of missing values in each column
missing_percentage = ((df.isnull().sum() / df.shape[0])*100).round(2)
print(missing_percentage)
```

```python
# Get the total number of missing values in the dataset
print(df.isnull().sum().sum())
```

```python
# Get the percentage of total missing values based on total data of dataset
miss_percentage = (df.isnull().sum().sum() / df.size) * 100
print(f"Missing value is {miss_percentage.round(2)}% of total data.")
```

```python
# Get the number of non missing values according to per column
print("Number of non missing values per column")
df.notnull().sum()
```

```python
# Get the percentage of non missing values according to per column
print("Percentage of non missing values per column")
((df.notnull().sum() / df.shape[0])*100).round(2)
```

```python
# Get the total percentage of non missing values per column
non_miss_perc = ((df.notnull().sum().sum() / df.size)*100).round(2)
print(f"Total non missing data is {non_miss_perc}%.")
```

```python
# Get the visualization of missing values
sns.heatmap(df.isnull())
plt.show()
```

```python
# Delete the "Credit_History" column as it has large number of missing values
df.drop(columns=["Credit_History"])
```

```python
# Get the type of data in column
df.info()
```

```python
# Fill the null value of "Gender" column using mode as it is categorical data type
df["Gender"] = df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
# Fill the null value of "Married" column using mode as it is categorical data type
df["Married"] = df["Married"].fillna(df["Married"].mode()[0])
```

```python
# Get the object data type in the dataset
df.select_dtypes(include="object")
```

```python
# Fill the null values of object data types with mode
for i in df.select_dtypes(include="object").columns:
    df[i] = df[i].fillna(df[i].mode()[0])
```

```python
# Get the numeric column in the dataset
from sklearn.impute import SimpleImputer # Import imputer from scikit learn
si = SimpleImputer(strategy="mean") # Use mean to fill null value
selected_col = df.select_dtypes(include="float").columns # Get the column with float data type
arr = si.fit_transform(df[selected_col]) # Apply the mean value to null data and create new series
new_df = pd.DataFrame(arr, columns=selected_col) # Create new dataframe with updated series
df = df.drop(selected_col, axis=1) # Drop the column with null value in float data type
df = pd.concat([df, new_df], axis=1) # Merge existing dataframe and new dataframe 
```

```python
# Implement One-hot encoding with with pandas
en_df = df[["Gender", "Married"]]
pd.get_dummies(en_df).head(2)
```

```python
# Implement One-Hot-Encoding with scikit-learn
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop="first")
arr = ohe.fit_transform(en_df).toarray()
new_en_df = pd.DataFrame(arr, columns=["Gender_Male", "Married_Yes"])
new_en_df.head(3)
```

## Implement Label Encoding

```python
# Create dataframe
samp_df = pd.DataFrame({"name": ["Cow", "Dog", "Cat", "Bird", "Cow"]})
```

```python
# Import label encoder and implement it to ordinal data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
samp_df["le_name"] = le.fit_transform(samp_df["name"])
samp_df
```

```python
# Implement label encoding to real dataset
le = LabelEncoder()
arr = le.fit(df["Property_Area"])
df["Property_Area"] = arr.transform(df["Property_Area"])
df.head(2)
```

```python
# Create new dataframe
samp_df = pd.DataFrame({"Size": ["s", "m", "l", "xl", "s", "m", "l", "s", "s", "l", "xl", "m"]})
samp_df.head(3)
```

```python
# Implement ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
```

```python
ord_data = [["s", "m", "l", "xl"]] # Create ordered data
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(samp_df[["Size"]])
```

```python
samp_df["En_Size"] = oe.transform(samp_df[["Size"]])
samp_df
```

```python
# Implement ordinal label encoding with map function
ord1 = {"s": 1, "m": 2, "l": 3, "xl": 4}
```

```python
samp_df["En_Size_1"] = samp_df["Size"].map(ord1)
samp_df
```

```python
# Calculate IQR
q1 = df["CoapplicantIncome"].quantile(0.25)
q3 = df["CoapplicantIncome"].quantile(0.75)
IQR = q3 - q1
min_range = q1 - (1.5 * IQR)
max_range = q3 + (1.5 * IQR)
print(min_range, max_range)
```

```python
# Draw boxplot
plt.figure(figsize=(15, 5))
sns.boxplot(x="CoapplicantIncome", data=df)
plt.show()
```

```python
new_df = df[df["CoapplicantIncome"] <= max_range]
new_df.head()
```

```python
new_df.shape
```

```python
sns.boxplot(x="CoapplicantIncome", data=new_df)
plt.show()
```

```python
# Calculate max and min range with z score
max_range = df["CoapplicantIncome"].mean() + (3*df["CoapplicantIncome"].std())
min_range = df["CoapplicantIncome"].mean() - (3*df["CoapplicantIncome"].std())
print(min_range, max_range)
```

```python
new_df = df[df["CoapplicantIncome"] <= max_range]
sns.boxplot(x="CoapplicantIncome", data=new_df)
plt.show()
```

```python
sns.distplot(new_df["CoapplicantIncome"])
plt.show()
```

```python
# Calculate z-score
z_score = (df["CoapplicantIncome"] - df["CoapplicantIncome"].mean()) / df["CoapplicantIncome"].std()
new_df = df.copy()
new_df["z_score"] = z_score
new_df = new_df[new_df["z_score"] < 3]
sns.boxplot(x="z_score", data=new_df)
```

## Using Transformer Function to convert into normal distribution

```python
q1 = df["CoapplicantIncome"].quantile(0.25)
q3 = df["CoapplicantIncome"].quantile(0.75)
iqr = q3 - q1
max_range = q3 + (1.5 * iqr)
min_range = q1 - (1.5 * iqr)
df = df[df["CoapplicantIncome"] <= max_range]
```

```python
df.shape
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(df[["CoapplicantIncome"]])
```

```python
arr = ft.transform(df[["CoapplicantIncome"]])
df["CoapplicantIncome_tf"] = arr
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Before")
sns.distplot(df["CoapplicantIncome"])

plt.subplot(1, 2, 2)
plt.title("After")
sns.distplot(df["CoapplicantIncome_tf"])

plt.show()
```
