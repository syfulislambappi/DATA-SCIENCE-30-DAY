## 1ï¸âƒ£ Import Libraries & Load Data

```
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("your_dataset.csv")  # Replace with actual file
df_raw = df.copy()

df.head()
```

## 2ï¸âƒ£ Initial Profiling & Quick Summary

```
df.info()
df.describe(include='all').T
```

```
# Missing value summary
df.isna().sum().sort_values(ascending=False)
```

## 3ï¸âƒ£ Data Type Fixing

```
# Example fixes â€“ adjust per dataset:

# Numeric conversion
numeric_cols = ['price','amount','score']  # CHANGE
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Date parsing
date_cols = ['date','signup_date','order_date']  # CHANGE
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

df.dtypes
```

## 4ï¸âƒ£ Text Normalization / Categorical Cleaning

```
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)

# Example: country normalization
if 'country' in df.columns:
    df['country'] = df['country'].str.upper().replace({
        'UNITED STATES':'US', 'U.S.':'US', 'USA':'US'
    })

df.head()
```

## 5ï¸âƒ£ Missing Value Handling

```
# Numeric: impute median
for col in df.select_dtypes(include=['float','int']).columns:
    df[col] = df[col].fillna(df[col].median())

# Categorical: impute 'Unknown'
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].replace(['nan','None'], np.nan).fillna('Unknown')

df.isna().sum().sort_values(ascending=False)
```

## 6ï¸âƒ£ Outlier Detection & Treatment

```
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    return data[(data[column] >= Q1 - 1.5*IQR) & (data[column] <= Q3 + 1.5*IQR)]

# Example â€“ customize:
numeric_cols = df.select_dtypes(include=['float','int']).columns
for col in numeric_cols:
    df = remove_outliers_iqr(df, col)

df.shape
```

## 7ï¸âƒ£ Duplicate Handling

```
df_before = df.shape[0]
df = df.drop_duplicates()
df_after = df.shape[0]
df_before, df_after
```

## 8ï¸âƒ£ Feature Engineering (Optional)

```
# Example: Extract day, month, year
if 'date' in df.columns:
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day

# Binning numeric values
# df['price_bin'] = pd.qcut(df['price'], 4, labels=False)

df.head()
```
