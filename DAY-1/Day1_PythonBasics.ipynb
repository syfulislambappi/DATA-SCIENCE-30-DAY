## 2. Loop Trhough Dictionary

```python
students = {
    "1": {"name": "Syful Islam", "mark": 67},
    "2": {"name": "Arafat Islam", "mark": 34},
    "3": {"name": "Nazmul Islam", "mark": 80},
    "4": {"name": "Mehzabin Chowdhury", "mark": 53},
    "5": {"name": "Arifa Islam", "mark": 40},
}

for student in students.values():
    if student["mark"] > 40:
        print(f"{student["name"]} passed and got {student["mark"]}.")
```

## 3. Check Prime Number

```python
number = int(input("Enter an integer number: "))

if number == 2:
    print(f"{number} is a prime number")

elif number > 2:
    remainder = []
    zeros = []
        
    for i in range(2, number):
        remainder.append(number % i)
        
    for i in remainder:
        if i == 0:
            zeros.append(i)

    if len(zeros) == 0:
        print(f"{number} is a prime number")
    else:
        print(f"{number} is not a prime number")
```
