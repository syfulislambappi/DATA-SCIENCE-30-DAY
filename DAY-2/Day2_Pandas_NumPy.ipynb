# NumPy Exercises

```python
# Create array from 1 to 50 and reshpae it (5, 10)
arr = np.arange(1, 51).reshape(5, 10)
print(arr)
```

```python
# Compute sum, mean, and standard deviation
sum = np.sum(arr)
mean = np.mean(arr)
std = np.std(arr)

print(f"Sum of the Array is {sum}")
print(f"Average of the Array is {mean}")
print(f"Standard Deviation of the Array is {std.round(2)}")
```

```python
# Slice first three rows
new_arr = arr[:3]

print(new_arr)
```

```python
# Multiply all values by two
new_arr = arr * 2

print(new_arr)
```
