# NumPy Exercises

```python
# Create array from 1 to 50 and reshpae it (5, 10)
arr = np.arange(1, 51).reshape(5, 10)
print(arr)
```

```python
# Compute sum, mean, and standard deviation
sum = np.sum(arr)
mean = np.mean(arr)
std = np.std(arr)

print(f"Sum of the Array is {sum}")
print(f"Average of the Array is {mean}")
print(f"Standard Deviation of the Array is {std.round(2)}")
```

```python
# Slice first three rows
new_arr = arr[:3]

print(new_arr)
```

```python
# Multiply all values by two
new_arr = arr * 2

print(new_arr)
```

# Pandas Exercises

```python
# Import pandas
import pandas as pd
```

```python
# Load csv file
df = pd.read_csv("titanic.csv")
```

```python
# Show the top five rows
df.head()
```

```python
# Show the shape of dataset
df.shape
```

```python
# Show the data types of each column
df.info()
```

```python
# Show the summary stats of each column
df.describe()
```

```python
# Get the column Age
df["Age"]
```

```python
# Get column age and fare
df[["Age", "Fare"]]
```

```python
# Show Passengers older than 30
df[df["Age"] > 30]
```

```python
# Show male passangers who paid more than 50
df[(df["Sex"] == "male") & (df["Fare"] > 50)]
```

```python
# Create new Column
df["AgeGroup"] = df["Age"].apply(lambda x: "Child" if x < 15 else "Adult")
df.head()
```

```python
# Drop Irrelevent column
df = df.drop(["SibSp", "Parch"], axis=1)
df
```

```python
# Sort the data by fare price
df.sort_values(by="Fare", ascending=False)
```

```python
# Show number of missing values in each column
df.isnull().sum()
```

```python
# Replace missing age with mean age
df["Age"] = df["Age"].fillna(df["Age"].mean())
df
```

```python
# Remove rows where embarked is missing
df = df[df["Embarked"].notnull()]
df
```

```python
# Insight one: number of passangers based on sex
df.groupby(["Sex"]).agg({
    "Sex": "count"
})
```

```python
# Insight two: Average age of Passangers
print(df["Age"].mean().round(2))
```

```python
# Insigt three: Average fare of Passangers
print(df["Fare"].mean().round(2))
```

```python
# Insight four: number of adult and child passangers
df.groupby("AgeGroup").agg({"AgeGroup": "count"})
```

```python
# Insight Five: Total number of passangers survived
print(df["Survived"].sum())
```
